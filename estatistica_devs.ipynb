```python
import pandas as pd
import matplotlib.pyplot as plt

# Dicionário de faturamento
dict_faturamento = {
    'data_ref': [
        '2023-01-01', '2020-02-01', '2021-03-01', '2022-04-01', '2023-05-01',
        '2023-06-01', '2020-07-01', '2021-08-01', '2022-09-01', '2023-10-01',
        '2022-11-01', '2023-12-01',
    ],
    'valor': [
        400000, 890000, 760000, 430000, 920000,
        340000, 800000, 500000, 200000, 900000,
        570000, 995000,
    ]
}

# Cria DataFrame
df = pd.DataFrame(dict_faturamento)

# Converte a coluna de datas para datetime
df['data_ref'] = pd.to_datetime(df['data_ref'])

# Ordena por data (opcional, mas geralmente útil para gráfico)
df = df.sort_values('data_ref')

# Cria gráfico de barras
plt.figure(figsize=(12, 6))
plt.bar(df['data_ref'].dt.strftime('%Y-%m'), df['valor'], color='skyblue')

# Adiciona título e rótulos
plt.title('Faturamento por Data de Referência')
plt.xlabel('Data de Referência')
plt.ylabel('Valor (R$)')
plt.xticks(rotation=45)
plt.tight_layout()

# Exibi gráfico
plt.show()
```
